```python
%load_ext autoreload
%autoreload 2
import edhec_risk_kit as erk
import pandas as pd
import numpy as np
from scipy.optimize import minimize
ind_return = erk.get_ind_returns()
ind_size = erk.get_ind_size()
ind_nfirms=erk.get_ind_nfirms()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap=ind_mktcap.sum(axis="columns")
```

```python
total_mktcap.plot(figsize=(12,6))
```

```python
total_mktcap
```

```python
ind_capweight= ind_mktcap.divide(total_mktcap,axis="rows")
```

```python
ind_capweight.head()
```

```python
ind_capweight[["Fin","Steel"]].plot(figsize=(16,8))
```

```python
total_market_return= (ind_capweight*ind_return).sum("columns")
```

```python
(ind_capweight*ind_return).head()
```

```python
total_market_return["2000":"2018"].plot()
```

```python
total_market_index= erk.drawdown(total_market_return).wealth
```

```python
total_market_index.plot()
```

```python
total_market_index['1980':].plot(figsize=(12,6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6)) #trailing 36 onth month moving average
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
tmi_tr36rets.plot(figsize=(12,6),label="Trailing 36 monthly return",legend=True)
##total_market_return.plot(label="Returns",legend=True)
```

### Rolling correlation- along with Multiindexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
#single row index
```

```python
# corr matrix is a multi index
```

```python
ts_corr.index.names
```

```python
ts_corr.index.names=['date','industry']
```

```python
ts_corr.tail()
```

```python
ind_tr36corr =ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())     
#ts_cor will be grouped by date, .apply gives a data frame that  
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="Trailing 36 month rets", legend=True ,figsize=(12,6))
ind_tr36corr.plot(label="Trailing 36 month corr", legend=True ,figsize=(12,6))
```
