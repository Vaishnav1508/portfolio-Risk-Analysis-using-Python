```python
safe_r[:] = 0.03/12
start = 1000
floor = 0.8
```

1. Cushion - asset value minus floor value 
2. compute an allocation to safe and risky assets ---> m*risk budget 
3. recompute the asset value based on the returns 

```python
def compound1(r):
    return (1+r).prod()-1

def compound2(r):
    return np.expm1(np.log1p(r).sum())
```
